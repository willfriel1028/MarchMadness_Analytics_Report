```python
import pandas as pd

safe = pd.read_csv("../data/safer_wins.csv")
agg = pd.read_csv("../data/aggressive_wins.csv")

del safe["Unnamed: 0"]
del agg["Unnamed: 0"]
```

```python
df = pd.merge(safe, agg, on=['Year', 'Team'], how='outer')

df.head()
```

```python
df["Diff"] = abs(df["Wins_Rank_Safer"] - df["Wins_Rank_Agg"])
```

```python
df.sort_values("Diff", ascending=False).head(25)
```

```python
for i in range(0,10):
    
    dif = df[df["Diff"] == i]
    print(i, ": ", len(dif))
```

```python
print("0: ", len(df[df["Diff"] == 0]))
print("1+: ", len(df[df["Diff"] != 0]))
```

```python
print("0: ", len(df[df["Diff"] == 0]))
print("1: ", len(df[df["Diff"] == 1]))
print("2+: ", len(df[df["Diff"] >= 2]))
```
